```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.cosmology import FlatLambdaCDM
```

```python
cc = FlatLambdaCDM(H0=100, Om0=0.315) 
def get_sigma_crit_inv(lzred, szred, cc=cc):
    if szred<lzred:
        return 0
    # some important constants for the sigma crit computations
    gee = 4.301e-9 #km^2 Mpc M_sun^-1 s^-2 gravitational constant
    cee = 3e5 #km s^-1
    # sigma_crit_calculations for a given lense-source pair
    Dang_l = cc.angular_diameter_distance(lzred).value
    Dang_s = cc.angular_diameter_distance(szred).value
    Dang_l_s = cc.angular_diameter_distance_z1z2(lzred,szred).value
    
    sigm_crit_inv =  Dang_l* Dang_l_s * 1.0/Dang_s
    #sigm_crit_inv =  Dang_l* Dang_l_s * (1+lzred)**2/Dang_s
    
    sigm_crit_inv = sigm_crit_inv * 4*np.pi*gee*1.0/cee**2
    #sigm_crit_inv = sigm_crit_inv #esd's are in pc not in Mpc


    #in comoving units
    inv_sig_cov = cc.comoving_distance(lzred).value*(cc.comoving_distance(szred).value - cc.comoving_distance(lzred).value)
    inv_sig_cov /=cc.comoving_distance(szred).value
    inv_sig_cov /=(1+lzred)
    inv_sig_cov *= 4*np.pi*gee*1.0/cee**2
    
 
    return sigm_crit_inv, inv_sig_cov
    
```

```python
zsrc = 0.8 #np.linspace(0.1,0.8,100)
zlen = np.linspace(0.0,0.8,100) 
yy =0.0 * zlen
yy1 =0.0 * zlen

for ii in range(len(zlen)):
    yy[ii] = get_sigma_crit_inv(zlen[ii], zsrc)[0]
    yy1[ii] = get_sigma_crit_inv(zlen[ii], zsrc)[1]
plt.subplot(2,2,1)
plt.plot(zlen, yy)
plt.plot(zlen, yy1)
print(yy/yy1)
plt.yscale('log')
plt.axvline(zsrc,color='k')
#plt.yscale('log')
plt.xlabel(r'$z_{\rm lens}$')
plt.ylabel(r'$\Sigma^{-1}_{\rm crit}(z_l, z_s = %2.2f)$'%zsrc)
```
